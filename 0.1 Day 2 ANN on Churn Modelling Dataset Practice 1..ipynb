```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('Churn_Modelling.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
# Lets drop the unwwanted column.
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Geography'] = le.fit_transform(df['Geography'])
```

```python
le2 = LabelEncoder()
df['Gender'] = le2.fit_transform(df['Gender'])
```

```python
df
```

```python
# Separating the Data into X and Y.
x = df.iloc[:,:-1]
x
```

```python
y = df.iloc[:,-1]
y 
```

```python
# Will performing deep learning model will always perform Scaling 
from sklearn.preprocessing import StandardScaler

# Create object of the class
sc = StandardScaler()

# Scaling the X Variavle which include Input Features 
# Remember we dont scale the output Variable
x = sc.fit_transform(x)
```

```python
x
```

```python
y = df['Exited'].values
y 
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size = 0.30, random_state = 1)
```

```python
# from here we start building Neural Network

import tensorflow as tf
from tensorflow.keras import Sequential # it is Used to build NN
from tensorflow.keras.layers import Dense # it is used to add layers
from sklearn.metrics import classification_report # Evaluation 
```

```python
# Building Neural Network

# Step 1 : initiate a model

ann = Sequential()

# Step 2: Add Layers to model 

ann.add(Dense(units = 12,activation = 'relu')) # Hidden Layers
ann.add(Dense(units = 1, activation = 'sigmoid')) # Output Layer

# step 3: Establishing connection with layers
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# step 4: fit the model
ann.fit(xtrain,ytrain,epochs = 100, batch_size = 32)

# Step 5: make predictions
ypred = ann.predict(xtest)
```

```python
ypred 
```

```python
# step 6: Set the Threshold Value 

ypred = np.where(ypred>0.5,1,0)
ypred
```

```python
print(classification_report(ytest,ypred))
```
