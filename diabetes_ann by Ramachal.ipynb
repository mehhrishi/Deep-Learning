### REplace Zero with Mean

```python
df['SkinThickness'].replace(0,df['SkinThickness'].mean(), inplace=True)
```

```python
df['BloodPressure'].replace(0,df['BloodPressure'].mean(), inplace=True)
```

```python
df['Glucose'].replace(0,df['Glucose'].mean(), inplace=True)
```

```python
df['Insulin'].replace(0,df['Insulin'].mean(), inplace=True)
```

```python
df['BMI'].replace(0,df['BMI'].mean(), inplace=True)
```

```python
df
```

```python
df.values
```

```python
x = df.drop(columns=["Outcome"])
```

```python
x
```

```python
#Use the StandardScaler to scale the data set
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x =sc.fit_transform(x)
```

```python
x
```

```python
y =df["Outcome"].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtest,xtrain,ytest,ytrain = train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
# from here we start building Neural Network

import tensorflow as tf
from tensorflow.keras import Sequential # it is Used to build NN
from tensorflow.keras.layers import Dense # it is used to add layers
from sklearn.metrics import classification_report # Evaluation 
```

```python
##Built a Model
#step 1-: init a model
ann=Sequential()
#step2-: add layers to model
ann.add(Dense(units=16,activation="relu")) #HL
ann.add(Dense(units=1,activation="sigmoid"))
#step3-: establish connection with layers
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
#step4-: fit the model
ann.fit(xtrain,ytrain,epochs=120,batch_size=32)
#step5-: make predictions
ypred=ann.predict(xtest)
```

```python
ann.history.history
```

```python
ypred
```

```python
#Step 6 Set a threshold
ypred = np.where(ypred>0.5,1,0)
ypred
```

```python
print(classification_report(ytest,ypred))
```
