### Replacing the value '0' (zero) for each column with their repesctive columns Mean Value.

```python
for i in df.iloc[:,1:6].columns:
    df[i].replace(0,df[i].mean(),inplace=True)
print('Done Replacing')
```

### Separating the Data into x and y for Training purpose

```python
x = df.iloc[:,:-1].values
```

```python
x
```

```python
y = df.iloc[:,-1].values
```

```python
y
```

### When you perform Deep learning so always Scale the data, here we are using StandardScalar.

```python
# Will performing deep learning model will always perform Scaling.
from sklearn.preprocessing import StandardScaler
```

```python
# Create object of the class
sc = StandardScaler()
```

```python
# Scaling the X Variavle which include Input Features 
# Remember we dont scale the output Variable
x = sc.fit_transform(x)
```

```python
x
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
xtrain
```

```python
xtrain.shape
```

```python
x.shape[0]*70/100
```

#### Lets build Ann Model.

```python
# from here we start building Neural Network

import tensorflow as tf # import the tensorflow Library
from tensorflow.keras import Sequential # it is used to build Neural Netwrok 
from tensorflow.keras.layers import Dense # it is used to Add Layers
from sklearn.metrics import classification_report # Evaluation.
```

```python
# Builing Neural Network

# step1: initiate a model
ann = Sequential()

# Step 2: Add Layers  to Model
ann.add(Dense(units = 12, activation = 'relu'))
# created a 1 Hidden Layer with 12 neurons with the Activation function called 'Hidden Layers'

# creating a Output Layer with 1 neuron and activation function will be = 'Sigmoid'
ann.add(Dense(units = 1, activation = 'sigmoid'))


# Step 3: Establishing connection with Layers
ann.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Step 4: fit the model
ann.fit(xtrain,ytrain, epochs = 120, validation_data = (xtest, ytest),verbose=1 ,batch_size = 32)
```

```python
model_loss = pd.DataFrame(ann.history.history)
model_loss
```

```python
model_loss.head()
```

```python
ann.history.history
```

```python
model_loss[['accuracy','val_accuracy']].plot()
plt.title('Accuracy Comaprision for Traing and Testing data')
```

```python
model_loss[['loss','val_loss']].plot()
plt.title('Error Comaprision for Traing and Testing data')
```

```python
#step5-: make predictions
ypred=ann.predict(xtest)
```

```python
ypred = np.where(ypred>0.5,1,0)
ypred
```

### Model Evaluation

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(ytest,ypred))
```
